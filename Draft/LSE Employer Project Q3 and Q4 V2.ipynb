### Overall where are the cyclist going to

```python
import pandas
from collections import Counter

plt.figure(figsize=(15,4))
letter_counts = Counter(Outer_London["Direction"])
df = pandas.DataFrame.from_dict(letter_counts, orient='index')
df.plot(kind='bar')

#To add plot labels
#plt.xlabel("Direction of Cyclists")
#plt.ylabel('Total Number of Cyclist')
#plt.title("Overview of cyclist Direction of outer London ")
```

### Which gender cycles more

```python
plt.figure(figsize=(15,4))
Outer_London[Outer_London.columns[10:12]].sum().plot.bar(rot=0, color='DarkTurquoise')
plt.xlabel("Gender of Cyclist")
plt.ylabel('Total Number of Cyclists')
plt.title("Gender Demographic Split of Cyclists in London")
plt.show()
```

```python
plt.figure(figsize=(15,4))
Outer_London[Outer_London.columns[10:15]].sum().plot.bar(rot=0, color='DarkTurquoise')
plt.xlabel("Gender of Cyclist")
plt.ylabel('Total Number of Cyclists')
plt.title("Gender Demographic Split of Cyclists in London")
plt.show()
```

### Are men and women cycling to different areas

```python
plt.figure(figsize=(15,4))
sns.lineplot(data=Outer_London, x = "Number of male cycles", y = "Total cycles", hue="Direction")
```

```python
plt.figure(figsize=(15,4))
sns.lineplot(data=Outer_London, x = "Number of female cycles", y = "Total cycles", hue="Direction")
```

### Do men and women cycle at different times

```python
plt.figure(figsize=(15,4))
# Plot a simple line chart
plt.plot(Outer_London['Start hour'], Outer_London['Number of male cycles'], color='g', label='Cycle Pattern of male cyclist in a day')

# Plot another line on the same chart/graph
plt.plot(Outer_London['Start hour'], Outer_London['Number of female cycles'], color='r', label='Cycle Pattern of female cyclist in a day')

plt.xlabel("Start hour")
plt.ylabel('Total Number of Cyclists')
plt.title("Aggregated Cycle Pattern in a day")
plt.legend()
plt.show()
```

### Are men and women cycling depending on per weather

```python
plt.figure(figsize=(15,4))
sns.scatterplot(x="Number of female cycles", y = "Total cycles", data = Outer_London, hue = "Weather", estimator=sum)
```

```python
plt.figure(figsize=(15,4))
sns.scatterplot(x="Number of male cycles", y = "Total cycles", data = Outer_London, hue = "Weather", estimator=sum)
```
